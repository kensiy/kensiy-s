```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras import layers
```

```python
# 加载数据
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# 查看数据集的大小
X_train.shape, y_train.shape
```

```python
'''数据预处理'''
# 归一化
X_train, X_test = X_train / 255.0, X_test / 255.0
# 将标签转化为独热编码
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
# 建立卷积神经网络模型
model = tf.keras.Sequential()
model.add(layers.Flatten(input_shape=(28,28))) # 输入层，展平，28x28
model.add(layers.Dense(128, activation='relu')) # 隐藏层
model.add(layers.Dense(10, activation='softmax')) # 输出层

model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['acc'])
```

```python
# 训练模型，batch_size设置为500，训练20轮
history = model.fit(X_train, y_train, batch_size=500, epochs=20)
```

```python
# 使用测试集进行测试
model.evaluate(X_test, y_test)
```

# 从结果可以看出，模型在训练集的准确率上达到99.31%，在测试集上的准确率达到97.87%，有较好的性能
